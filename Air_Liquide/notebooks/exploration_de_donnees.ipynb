```python
# Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Chargement des données
data = pd.read_csv("../data/air_liquid_data.csv", index_col='Date')
print(data.columns)
data=data[['Low','Volume']] #je fais le choix de travailler que sur la valeur minimale de chaque journée
data.head()
```

```python
data.shape
```

```python
data.columns = ['Price','Volume']
```

```python
fs = (6,3)

# Exploration de la distribution des prix
plt.figure(figsize=fs)
sns.histplot(data=data, x="Price", kde=True)
plt.title("Distribution des prix")
plt.xlabel("Prix")
plt.show()

# Exploration de la distribution du volume
plt.figure(figsize=fs)
sns.histplot(data=data, x="Volume", kde=True)
plt.title("Distribution du volume")
plt.xlabel("Volume")
plt.show()

# Exploration de la relation entre le prix et le volume
plt.figure(figsize=fs)
sns.scatterplot(data=data, x="Volume", y="Price")
plt.title("Relation entre le prix et le volume")
plt.xlabel("Volume")
plt.ylabel("Prix")
plt.show()

# Représentation série temporelle (date en abscisse)
data[['Price']].plot(figsize=(12,8))
```
